```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join('..', 'src')))

import pandas as pd
from visualization import filter_words_by_tfidf, generate_wordcloud

# 前処理データロード
df = pd.read_csv('../data/preprocessed_data.csv')

# raw_calculate_tfidfカラムを辞書形式に変換
# もし辞書形式で読み込まれていない場合は、eval() で変換
df['raw_calculate_tfidf'] = df['raw_calculate_tfidf'].apply(eval)

# しきい値を設定してTF-IDFによる単語をフィルタ
threshold = 0.4
filtered_words = filter_words_by_tfidf(df['raw_calculate_tfidf'], threshold)

# フォントのパスを指定
font_path = "../data/ipaexg.ttf"

# フィルタされた単語を使ってワードクラウドを生成（フォント指定を追加）
generate_wordcloud(filtered_words, font_path=font_path)
```
